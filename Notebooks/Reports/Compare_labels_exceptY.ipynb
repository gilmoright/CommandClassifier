```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
REPORTS_DIR = "/s/ls4/users/grartem/RL_robots/CommandClassifier/models/"
```

```python
# AVG macro for every label except Y
table = []
for experimentName in ["rubert_tiny_multilabel", "rubert_tiny2_multilabel", "rubert_tiny2_multilabel_useY", "MyMultiTiny2"]:
    table.append({
            "experiment": experimentName,
        })
    for fold_i in range(5):
        if not os.path.exists(os.path.join(REPORTS_DIR, experimentName, "fold_{}".format(fold_i), "reports", "epoch-{}_avg_report.json".format(10))):
            continue
        with open(os.path.join(REPORTS_DIR, experimentName, "fold_{}".format(fold_i), "reports", "epoch-{}_avg_report.json".format(10)), "r") as f:
            report = json.load(f)
        counter, avg = 0, 0
        for k, v in report.items():
            if k in ['correct_samples_perc', 'avg_macro_f1', 'avg_acc', "[y]_macrof1"]:
                continue
            if "_acc" in k:
                continue
            avg += v
            counter += 1
        table[-1]["fold_{}".format(fold_i)] = avg / counter
table_df = pd.DataFrame(table)
table_df
```
